```python
import pandas as pd
import os
```

```python
dfs = []
def extract_name_csvs(csv_file, dfs):
    df = pd.read_csv(f'inputs/names/{csv_file}')
    dfs.append(df)

csv_files = [f for f in os.listdir('inputs/names') if f.endswith(".csv")]
for idx, csv_file in enumerate(csv_files):
    extract_name_csvs(csv_file, dfs)
df_name = pd.concat([dfs[0], dfs[1], dfs[2]])
```
